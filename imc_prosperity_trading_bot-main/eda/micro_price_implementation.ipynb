```python
def get_df(df):
    df = df[['timestamp', 'product', 'mid_price', 'bid_price_1', 'ask_price_1', 'bid_volume_1', 'ask_volume_1']].copy()
    df['mid_price'] = df['mid_price'].astype(float)
    df['bid_price_1'] = df['bid_price_1'].astype(float)
    df['ask_price_1'] = df['ask_price_1'].astype(float)
    df['bid_volume_1'] = df['bid_volume_1'].astype(float)
    df['ask_volume_1'] = df['ask_volume_1'].astype(float)

    df['imb'] = df['bid_volume_1'].astype(float)/(df['bid_volume_1'].astype(float)+df['ask_volume_1'].astype(float))
    df['wmid'] = df['ask_price_1'].astype(float)*df['imb']+df['bid_price_1'].astype(float)*(1-df['imb'])
    # df.rename(columns = {'mid_price':'mid', 'bid_price_1':'bid', 'ask_price_1':'ask', 'bid_volume_1':'bs', 'ask_volume_1':'as','timestamp':'time'}, inplace = True)
    return df.reset_index(drop = True)
```

```python
def prep_data_sym(T, n_imb, dt, n_spread):
    # Rename for compatibility with original function
    T = T.copy()
    T['bid'] = T['bid_price_1']
    T['ask'] = T['ask_price_1']
    T['bs'] = T['bid_volume_1']
    T['as'] = T['ask_volume_1']
    T['time'] = T['timestamp']  # Optional: modify if needed

    # Compute spread and tick size
    spread = T['ask'] - T['bid']
    ticksize = np.round(min(spread[spread > 0]) * 100) / 100

    # Add spread and mid price
    T['spread'] = np.round((T['ask'] - T['bid']) / ticksize) * ticksize
    T['mid'] = (T['bid'] + T['ask']) / 2

    # Filter out large spreads
    T = T.loc[(T['spread'] <= n_spread * ticksize) & (T['spread'] > 0)]

    # Compute imbalance and discretize
    T['imb'] = T['bs'] / (T['bs'] + T['as'])
    T['imb_bucket'] = pd.qcut(T['imb'], n_imb, labels=False, duplicates='drop')
    # Step ahead variables
    T['next_mid'] = T['mid'].shift(-dt)
    T['next_spread'] = T['spread'].shift(-dt)
    T['next_time'] = T['time'].shift(-dt)
    T['next_imb_bucket'] = T['imb_bucket'].shift(-dt)

    # Change in mid price
    T['dM'] = np.round((T['next_mid'] - T['mid']) / ticksize * 2) * ticksize / 2

    # Filter abnormal price changes
    T = T.loc[(T['dM'] <= ticksize * 1.1) & (T['dM'] >= -ticksize * 1.1)]

    # Symmetrize
    T2 = T.copy(deep=True)
    T2['imb_bucket'] = n_imb - 1 - T2['imb_bucket']
    T2['next_imb_bucket'] = n_imb - 1 - T2['next_imb_bucket']
    T2['dM'] = -T2['dM']
    T2['mid'] = -T2['mid']

    T3 = pd.concat([T, T2])
    T3.index = pd.RangeIndex(len(T3.index))

    return T3, ticksize
```

```python
def estimate(T, n_imb, n_spread):
    # Q matrix: transition counts where dM == 0 (no price move)
    no_move = T[T['dM'] == 0]
    no_move_counts = no_move.pivot_table(
        index=['next_imb_bucket'],
        columns=['spread', 'imb_bucket'],
        values='time',
        fill_value=0,
        aggfunc='count'
    ).unstack().fillna(0)

    # Convert pivot table to matrix format
    Q_counts = no_move_counts.values[:n_imb * n_imb].reshape(n_imb, n_imb)

    # Add more blocks if multiple spreads exist
    for i in range(1, n_spread):
        start = i * n_imb * n_imb
        end = (i + 1) * n_imb * n_imb
        if end > len(no_move_counts):
            break
        Qi = no_move_counts.values[start:end].reshape(n_imb, n_imb)
        Q_counts = block_diag(Q_counts, Qi)

    # R1 matrix: transitions where dM ≠ 0 (price moved)
    move = T[T['dM'] != 0]
    move_counts = move.pivot_table(
        index=['dM'],
        columns=['spread', 'imb_bucket'],
        values='time',
        fill_value=0,
        aggfunc='count'
    ).unstack().fillna(0)

    R_counts = move_counts.values.reshape(n_imb * n_spread, 4)

    # Normalize transition matrix (Q + R1)
    T1 = np.concatenate((Q_counts, R_counts), axis=1).astype(float)
    for i in range(T1.shape[0]):
        row_sum = T1[i].sum()
        if row_sum > 0:
            T1[i] = T1[i] / row_sum

    Q = T1[:, :n_imb * n_spread]
    R1 = T1[:, n_imb * n_spread:]

    # Reward vector
    K = np.array([-0.01, -0.005, 0.005, 0.01])

    # R2 matrix: (next_spread, next_imb_bucket) transitions for spread/imb states
    move_counts_2 = move.pivot_table(
        index=['spread', 'imb_bucket'],
        columns=['next_spread', 'next_imb_bucket'],
        values='time',
        fill_value=0,
        aggfunc='count'
    ).fillna(0)

    R2_counts = move_counts_2.values.reshape(n_imb * n_spread, n_imb * n_spread)

    T2 = np.concatenate((Q_counts, R2_counts), axis=1).astype(float)
    for i in range(T2.shape[0]):
        row_sum = T2[i].sum()
        if row_sum > 0:
            T2[i] = T2[i] / row_sum

    Q2 = T2[:, :n_imb * n_spread]
    R2 = T2[:, n_imb * n_spread:]

    # Expected gain vector (G1) and transition matrix (B)
    I = np.eye(n_imb * n_spread)
    G1 = np.dot(np.dot(np.linalg.inv(I - Q), R1), K)
    B = np.dot(np.linalg.inv(I - Q), R2)

    return G1, B, Q, Q2, R1, R2, K
```

```python
def plot_Gstar(ticker, G1, B, T, n_imb=10, n_spread=2):
    # Compute long-term expected gain G*
    G = G1.copy()
    for _ in range(5):
        G = G + B @ G

    # x-axis: imbalance bucket centers
    imb_bucket_centers = np.linspace(0, 1, n_imb)

    plt.figure(figsize=(10, 6))
    
    # Optional baseline comparisons (dashed lines)
    plt.plot(imb_bucket_centers, np.zeros(n_imb), label='Mid adj (0)', linestyle='--', color='gray')
    plt.plot(imb_bucket_centers, np.linspace(-0.005, 0.005, n_imb), label='Weighted mid adj', linestyle='--', color='lightblue')

    # Plot each spread level
    for i in range(n_spread):
        adj = G[i * n_imb : (i + 1) * n_imb]
        plt.plot(imb_bucket_centers, adj, label=f"Spread = {i + 1} tick adj", marker='o')

    plt.ylim(-0.005, 0.005)
    plt.xlabel("Order Book Imbalance")
    plt.ylabel("Expected Price Adjustment")
    plt.title(f"{ticker} - Long-term Gain G* by Spread & Imbalance")
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.show()

    return G
```
